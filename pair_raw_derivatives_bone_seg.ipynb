```python
import json
from pathlib import Path

root = Path("/vol/miltank/projects/practical_sose25/atlas_based_registeration/data/dataset-myelom")
rawdata_root = root / "rawdata"
deriv_root = root / "derivatives"

pairs = {}
index = 1

for subject_dir in rawdata_root.glob("CTFU*"):
    subject = subject_dir.name
    for ses_dir in subject_dir.glob("ses-*"):
        session = ses_dir.name
        for raw_file in ses_dir.glob("sub-*_ct.nii.gz"):
            seg_filename = raw_file.name.replace("_ct.nii.gz", "_seg-bone_msk.nii.gz")
            seg_file = deriv_root / subject / session / seg_filename

            if seg_file.exists():
                pairs[index] = [str(raw_file), str(seg_file)]
                index += 1
            else:
                print(f"no seg for: {raw_file.name}")

# Speichern
with open("/vol/miltank/users/fratzke/Documents/atlas-based-registration/results/paired_raw_seg.json", "w") as f:
    json.dump(pairs, f, indent=4)

print(f"{index - 1} valid pairs found and saved.")
```
